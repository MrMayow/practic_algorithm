```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import detrend, butter, filtfilt, freqz, firwin, lfilter
import scipy.io
import pandas as pd
```

```python
SAMPLING_RATE = 9800 # Частота дискретизации
FILE_PATH= "normal/87988-498701.csv" # Пусть до .csv файла с данными
```

```python
def plot_signal_spectrum(signal, sampling_rate, min_x = 0.5, max_x = 30):
    n = len(signal)
    freq = np.fft.fftfreq(n, d=1/sampling_rate)
    fft_values = np.fft.fft(signal)
    magnitude = np.abs(fft_values)
    half_n = n // 2
    freq = freq[:half_n]
    magnitude = magnitude[:half_n]
    plt.figure(figsize=(10, 6))
    plt.plot(freq, np.log10(magnitude))
    plt.xlim([0, 200])
    #plt.ylim([-3, 3])
    plt.title('Signal Spectrum')
    plt.xlabel('Frequency (Hz)')
    plt.ylabel('Magnitude')
    plt.grid(True)
    plt.show()
```

```python
df = pd.read_csv(FILE_PATH)
values = df["vibration"].values
plot_signal_spectrum(values, SAMPLING_RATE, 0.5, 4900)
```
