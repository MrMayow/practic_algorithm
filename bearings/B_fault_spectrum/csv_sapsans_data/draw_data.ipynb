```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python


# Загрузка данных из CSV-файла
df = pd.read_csv('defect/3545353-3662598.csv')  # замените на имя вашего файла

# Укажите вашу частоту дискретизации (например, 1000 Гц)
sampling_rate = 9800  # Гц

# Создаём временную ось
time = np.arange(len(df)) / sampling_rate

# Строим график
plt.figure(figsize=(10, 5))
plt.plot(time, df.iloc[:, 0])
plt.xlabel('Время, сек')
plt.ylabel('Значение')
plt.title('')
plt.grid(True, linestyle='--', linewidth=0.5)
plt.show()
```
